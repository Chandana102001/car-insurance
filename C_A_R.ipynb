```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car = pd.read_excel('Car features.xlsx')
car
```

```python
# Extract max_torque_Nm, max_torque_rpm, max_power_bhp, max_power_rpm
car['max_torque_Nm'] = car['max_torque'].str.extract(r'(\d+)Nm').astype(float)
car['max_torque_rpm'] = car['max_torque'].str.extract(r'@(\d+)rpm').astype(float)
car['max_power_bhp'] = car['max_power'].str.extract(r'([\d.]+)bhp').astype(float)
car['max_power_rpm'] = car['max_power'].str.extract(r'@(\d+)rpm').astype(float)

# Drop the original columns
car.drop(['max_torque', 'max_power'], axis=1, inplace=True)

# Display the resulting DataFrame
car
```

```python
# List all column names:
car.columns.tolist()
```

```python
# Define the desired column order
my_order = ['policy_id', 'make', 'segment', 'model', 'max_torque_Nm', 'max_torque_rpm', 'max_power_bhp', 'max_power_rpm',
            'fuel_type', 'engine_type', 'airbags', 'is_esc', 'steering_type', 'is_adjustable_steering', 'is_tpms', 
            'is_parking_sensors', 'is_parking_camera', 'rear_brakes_type', 'is_brake_assist', 'displacement', 'cylinder', 
            'transmission_type', 'gear_box', 'turning_radius', 'length', 'width', 'height', 'gross_weight', 
            'is_front_fog_lights', 'is_rear_window_wiper', 'is_rear_window_washer', 'is_rear_window_defogger',  
            'is_power_door_locks',  'is_central_locking', 'is_power_steering', 'is_driver_seat_height_adjustable', 
            'is_day_night_rear_view_mirror', 'is_ecw', 'is_speed_alert', 'ncap_rating']
# Rearrange columns
car = car[my_order]
car
```

```python
# Display basic information about the dataset:
car.info()
```

```python
# Many of the columns depend on the car model. So, sorting the data by 'model':
car.sort_values(by='model', inplace=True)
car
```

```python
# Extract the numeric part of the 'model' column and convert to integers
car['model_numeric'] = car['model'].str.extract('(\d+)').astype(int)
# Sort the DataFrame based on both the numeric and string parts of 'model'
car_sorted = car.sort_values(by=['model_numeric', 'model'], ascending=[True, True])
# Drop the temporary 'model_numeric' column if you don't need it
car_sorted = car_sorted.drop(columns=['model_numeric'])
# Display the sorted DataFrame
car_sorted
```

```python
# Checking for missing values:
car_sorted.isnull().sum()
```

```python
# There are many missing values in the following columns: 
# max_torque_Nm                        7
# max_torque_rpm                       7
# max_power_bhp                        3
# max_power_rpm                        3
# fuel_type                           26
# engine_type                          3
# airbags                              9
# is_esc                               3
# steering_type                        0
# is_adjustable_steering               7
# is_tpms                              3
# is_parking_sensors                   7
# is_parking_camera                    3
# rear_brakes_type                     0
# is_brake_assist                      0
# displacement                         7
# cylinder                             7
# All these depend on the car model. So, we can fill the missing datas on the basis of the car model:

car_sorted['fuel_type'] = car_sorted.groupby('model')['fuel_type'].ffill()
car_sorted['max_torque_Nm'] = car_sorted.groupby('model')['max_torque_Nm'].ffill()
car_sorted['max_torque_rpm'] = car_sorted.groupby('model')['max_torque_rpm'].ffill()
car_sorted['max_power_bhp'] = car_sorted.groupby('model')['max_power_bhp'].ffill()
car_sorted['max_power_rpm'] = car_sorted.groupby('model')['max_power_rpm'].ffill()
car_sorted['engine_type'] = car_sorted.groupby('model')['engine_type'].ffill()
car_sorted['airbags'] = car_sorted.groupby('model')['airbags'].ffill()
car_sorted['is_adjustable_steering'] = car_sorted.groupby('model')['is_adjustable_steering'].ffill()
car_sorted['is_parking_camera'] = car_sorted.groupby('model')['is_parking_camera'].ffill()
car_sorted['is_parking_sensors'] = car_sorted.groupby('model')['is_parking_sensors'].ffill()
car_sorted['cylinder'] = car_sorted.groupby('model')['cylinder'].ffill()
car_sorted['displacement'] = car_sorted.groupby('model')['displacement'].ffill()
car_sorted['is_tpms'] = car_sorted.groupby('model')['is_tpms'].ffill()
car_sorted['is_esc'] = car_sorted.groupby('model')['is_esc'].ffill()
car_sorted
```

```python
# Checking for missing values:
car_sorted.isnull().sum()
```

```python
# Changing yes and no data to 1 and zero so that numerical analysis becomes easy:
# Replace 'Yes' with 1 and 'No' with 0:
car_sorted['is_esc'] = car_sorted['is_esc'].replace({'Yes': 1, 'No': 0})
car_sorted['is_adjustable_steering'] = car_sorted['is_adjustable_steering'].replace({'Yes': 1, 'No': 0})
car_sorted['is_tpms'] = car_sorted['is_tpms'].replace({'Yes': 1, 'No': 0})
car_sorted['is_parking_sensors'] = car_sorted['is_parking_sensors'].replace({'Yes': 1, 'No': 0})
car_sorted['is_parking_camera'] = car_sorted['is_parking_camera'].replace({'Yes': 1, 'No': 0})
car_sorted['is_front_fog_lights'] = car_sorted['is_front_fog_lights'].replace({'Yes': 1, 'No': 0})
car_sorted['is_rear_window_wiper'] = car_sorted['is_rear_window_wiper'].replace({'Yes': 1, 'No': 0})
car_sorted['is_rear_window_washer'] = car_sorted['is_rear_window_washer'].replace({'Yes': 1, 'No': 0})
car_sorted['is_rear_window_defogger'] = car_sorted['is_rear_window_defogger'].replace({'Yes': 1, 'No': 0})
car_sorted['is_brake_assist'] = car_sorted['is_brake_assist'].replace({'Yes': 1, 'No': 0})
car_sorted['is_power_door_locks'] = car_sorted['is_power_door_locks'].replace({'Yes': 1, 'No': 0})
car_sorted['is_central_locking'] = car_sorted['is_central_locking'].replace({'Yes': 1, 'No': 0})
car_sorted['is_power_steering'] = car_sorted['is_power_steering'].replace({'Yes': 1, 'No': 0})
car_sorted['is_driver_seat_height_adjustable'] = car_sorted['is_driver_seat_height_adjustable'].replace({'Yes': 1, 'No': 0})
car_sorted['is_day_night_rear_view_mirror'] = car_sorted['is_day_night_rear_view_mirror'].replace({'Yes': 1, 'No': 0})
car_sorted['is_ecw'] = car_sorted['is_ecw'].replace({'Yes': 1, 'No': 0})
car_sorted['is_speed_alert'] = car_sorted['is_speed_alert'].replace({'Yes': 1, 'No': 0})

car_sorted
```
